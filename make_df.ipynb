## 삼성의 2013~2014 co-inventor 특허 데이터 추출

```python
#applocation_sample에서 2013~2014 특허 데이터 추출
select_date_df = application_sample[application_sample['date'].str.contains('2013|2014', na=False)]
```

```python
#rawassignee_df에서 2013-2014 특허 데이터 추출
select_rawassignee_df = rawassignee_df[(rawassignee_df['patent_id'].isin(select_date_df['patent_id']))==True]

#rawassignee_df의 2013-2014 특허중 삼성의 특허만 뽑음
select_rawassignee_df = select_rawassignee_df[select_rawassignee_df['organization'].str.contains('Samsung', na=False)]
```

```python
#rawinventor_df에서 삼성 특허만 뽑음
samsung_p_df = rawinventor_df[(rawinventor_df['patent_id'].isin(select_rawassignee_df['patent_id']))==True]
```

```python
#공동 저자 수 데이터
num_sam_co_df = samsung_p_df[['patent_id','inventor_id']].groupby('patent_id').size().reset_index(name='num_co_inventor')

#공동 저자 수가 2명 이상인 특허 데이터만 남김
num_sam_co_df = num_sam_co_df[num_sam_co_df["num_co_inventor"] > 1]
```

```python
#공동 저자가 있는 삼성의 특허 데이터
sam_co_p_df = samsung_p_df[(samsung_p_df['patent_id'].isin(num_sam_co_df['patent_id']))==True]

#name_first와 name_last를 합쳐서 inventor_name 생성
sam_co_p_df['inventor_name'] = sam_co_p_df.name_first.astype(str).str.cat(sam_co_p_df.name_last.astype(str), sep=' ')
```

```python
drop_list = ['uuid', 'rawlocation_id', 'name_first', 'name_last', 'sequence']
fin_samsung_df = sam_co_p_df.drop(drop_list, 1).sort_values('patent_id').reset_index().drop('index', 1)
```

## inventor_id로 patent_id, astract 찾는 데이터 생성

```python
#rawinventor와 application_sample은 patent_id로 연결됨
#application_sample과 application_update은 number로 연결됨
```

```python
application_df = application_update.drop('number',1)
application_df['number'] = application_update['number_updated'].apply(lambda x : str(x)[:-1])
```

```python
merge_df = pd.merge(rawinventor_df, application_sample, on='patent_id')
merge_df = pd.merge(merge_df, application_df, on='number')
```

```python
merge_df_col = set(merge_df.columns.values)
use_col = {'patent_id','abstract'}

drop_col_list = list(merge_df_col - use_col)
```

```python
fin_absract_df = merge_df.drop(drop_col_list,1).drop_duplicates(subset='patent_id', keep='first')
```
